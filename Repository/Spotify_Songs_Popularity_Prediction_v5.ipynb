# HANDLING SKEWNESS

```python
# Function to plot skewness for specified columns
def plot_skewness(df, columns, title_suffix):
    plt.figure(figsize=(12, len(columns) * 4))  # Adjust the figure size
    for i, col in enumerate(columns):
        plt.subplot(len(columns), 2, i + 1)
        sns.histplot(df[col], bins=20, kde=True)  # Histogram with KDE
        plt.title(f'{col} Distribution {title_suffix}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()
```

```python
from scipy import stats

# List of skewed columns (We have taken columns only which are having >= +1 or <= -1 )
skewed_cols = ['speechiness', 'acousticness', 'instrumentalness', 'liveness', 'duration_ms', 'loudness']

plot_skewness(df, skewed_cols, 'Before Transformation')

# Shifting each column to make all values positive
for col in skewed_cols:
    shift = abs(df[col].min()) + 1  # Calculating the shift amount
    df[col] = df[col] + shift       # Shifting values in the column to be positive
```

```python
#Detecting and caping outliers using IQR method
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df
```

```python
# Applying IQR capping for the skewed columns
for col in skewed_cols:
    df = cap_outliers_iqr(df, col)
```

```python
#Applying Box-Cox transformation to reduce skewness
from sklearn.preprocessing import PowerTransformer

box_cox_transformer = PowerTransformer(method='box-cox')
df[skewed_cols] = box_cox_transformer.fit_transform(df[skewed_cols])
```

```python
# Verifying the skewness after Box-Cox transformation
post_boxcox_skewness = df[skewed_cols].skew()
print("Skewness after Box-Cox transformation:")
print(post_boxcox_skewness)
plot_skewness(df, skewed_cols, 'After Transformation')
```
