# HANDLING SKEWNESS

```python
from scipy import stats

# List of skewed columns (We have taken columns only which are having >= +1 or <= -1 )
skewed_cols = ['speechiness', 'acousticness', 'instrumentalness', 'liveness', 'duration_ms', 'loudness']

# Shifting each column to make all values positive
for col in skewed_cols:
    shift = abs(df[col].min()) + 1  # Calculating the shift amount
    df[col] = df[col] + shift       # Shifting values in the column to be positive
```

```python
#Detecting and caping outliers using IQR method
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df
```

```python
# Applying IQR capping for the skewed columns
for col in skewed_cols:
    df = cap_outliers_iqr(df, col)
```

```python
#Applying Box-Cox transformation to reduce skewness
from sklearn.preprocessing import PowerTransformer

box_cox_transformer = PowerTransformer(method='box-cox')
df[skewed_cols] = box_cox_transformer.fit_transform(df[skewed_cols])
```

```python
# Verifying the skewness after Box-Cox transformation
post_boxcox_skewness = df[skewed_cols].skew()
print("Skewness after Box-Cox transformation:")
print(post_boxcox_skewness)
```

# Linear regression

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python

df_sorted = spotify_data_cleaned.sort_values(by='track_id')

sample_data = df_sorted.head(15000)
sample_data
```

```python
###SAMPLE DATA ####

#Defining features (X) and target (y)
X = sample_data.drop(columns=['track_popularity'])
y = sample_data['track_popularity'] 

#Checking for the non numeric columns
non_numeric_columns = X.select_dtypes(include=['object']).columns

# Encode non-numeric columns
X = pd.get_dummies(X, columns=non_numeric_columns, drop_first=True) 

X_sample = X.sample(frac=0.5, random_state=42)
y_sample = y.loc[X_sample.index]

#Splitting the data into Training & Testing dataset
X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, random_state=42)

#Setting up k-fold cross-validation on the training data alone
kf = KFold(n_splits=5, shuffle=True, random_state=42)  # 5-fold cross-validation

```

```python
from sklearn.preprocessing import StandardScaler
#Linear regression models can be sensitive to the scale of the input data. Scaling the features (standardization or normalization) often resolves numerical issues.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Initializing the LinearRegression model which works as baseline prediction for our project
model = LinearRegression()
```

```python
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
```

```python
#Performing cross-validation on the training set only
cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=kf, scoring='neg_mean_squared_error',n_jobs=-1)
cv_rmse = np.sqrt(-cv_scores)  # Convert negative MSE to RMSE

#Displaying the cross-validation RMSE for each fold and the mean RMSE
print("Cross-Validation RMSE for each fold:", cv_rmse)
print("Mean Cross-Validation RMSE:", cv_rmse.mean())
```

```python
#Training the model on the full training set and evaluate on the test set
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
```

```python
#Calculating the metrics on the test set
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
test_mae = mean_absolute_error(y_test, y_pred)
test_mse = mean_squared_error(y_test, y_pred)
print("Test Set RMSE:", test_rmse)
print("Test Set MAE:", test_mae)
print("Test Set MSE:", test_mse)
```
