```python
plt.style.use('seaborn-v0_8-pastel')
```

```python
# song popularity histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['track_popularity'], bins=20, kde=True)

# Add title and labels
plt.title('Song Popularity Distribution with KDE')
plt.xlabel('Popularity')

# Show the plot
plt.show()

# track_popularity sknewness is -0.233485. Also from the Histogran, it is left skewed.
```

```python
# track_popularity is left skewed (-0.233485)
# checking for outliners
# Plot Demand outliers
fig, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(data=df, y='track_popularity', ax=ax)
ax.set_title('Showing outliers within Track Popularity')
plt.tight_layout()
plt.show()
# No outliers
```

```python
# Box plot of popularity by genre
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='playlist_genre', y='track_popularity')
plt.xticks(rotation=45)
plt.title('Box Plot of Popularity by Genre')
plt.ylabel('Popularity')
plt.xlabel('Genre')
plt.show()
```

```python
features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']

for feature in features:
    plt.figure(figsize=(5, 5))
    plt.scatter(df[feature], df['track_popularity'], alpha=0.5)
    plt.title(f'Scatter Plot of {feature.capitalize()} vs. Popularity')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Popularity')
    plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Violin plot of popularity by genre
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='playlist_genre', y='track_popularity')
plt.xticks(rotation=45)
plt.title('Violin Plot of Popularity by Genre')
plt.ylabel('Popularity')
plt.xlabel('Genre')
plt.show()
```

```python
# Pair plot of selected features
sns.pairplot(df[['track_popularity', 'danceability', 'energy', 'tempo', 'loudness']])
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python
# Bar plot of average popularity by artist
average_popularity = df.groupby('track_artist')['track_popularity'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
average_popularity.plot(kind='bar', edgecolor='black')
plt.title('Top 10 Artists by Average Popularity')
plt.xlabel('Artist')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45)
plt.show()
```
